```python
import pandas as pd
import numpy as np
from nanoid import generate

df = pd.read_csv('atus_2019.csv')
```

```python
def activity_processor(activity):
    if activity == 10101:
        return 'Sleeping'
    elif activity in range(10000, 20000):
        return 'Personal Care'
    elif activity in range(20000, 30000):
        return 'Housework'
    elif activity in range(30000, 40000):
        return 'Household Care'
    elif activity in range(40000, 50000):
        return 'Non-Household Care'
    elif activity in range(50000, 60000):
        return 'Work'
    elif activity in range(60000, 70000):
        return 'Education'
    elif activity in range(70000, 80000):
        return 'Shopping'
    elif activity in range(80000, 90000):
        return 'Pro. Care Services'
    elif activity in range(90000, 100000):
        return 'Household Services'
    elif activity in range(100000, 110000):
        return 'Government Services'
    elif activity in range(110000, 120000):
        return 'Eating and Drinking'
    elif activity in range(120000, 130000):
        return 'Leisure'
    elif activity in range(130000, 140000):
        return 'Sports'
    elif activity in range(140000, 150000):
        return 'Religion'
    elif activity in range(150000, 160000):
        return 'Volunteering'
    elif activity in range(160000, 170000):
        return 'Phone Calls'
    elif activity in range(180000, 190000):
        return 'Traveling'
    else:
        return 'Miscellaneous'
```

```python
df_actual = df[["WT20", "ACTIVITY", "DURATION"]]
df_actual.loc[:, "ACTIVITY"] = df_actual.ACTIVITY.apply(activity_processor) 
df_actual.head(20)
```

```python
new_data = {};
for index, row in df.iterrows():
    personID = str(round(row["WT20"])).strip("0") + str(round(row["CASEID"])).rstrip("0")
    if personID not in new_data:
        for x in range(row["DURATION"]):
            if x == 0:
#                 print("person not in dict adding them")
                new_data[personID] = [activity_processor(row["ACTIVITY"])]
            else:
                new_data[personID].append(activity_processor(row["ACTIVITY"]))
    else:
#         print("adding value to person")
        for x in range(row["DURATION"]):
            if (len(new_data[personID]) < 1440):
                new_data[personID].append(activity_processor(row["ACTIVITY"]))
    
```

```python
df_processed = pd.DataFrame.from_dict(new_data, orient ='index') 
# df_processed.head(20)
df_unique_activites = df_processed[240].unique()
df_unique_activites
```

```python
initial_df  = pd.crosstab(index=df_processed[0], columns='count').reindex(index=df_unique_activites, fill_value=0)
initial_df_sum = initial_df/initial_df.sum()
newdf = initial_df_sum['count']
# pd.DataFrame(newdf)['count']
initial_prob_array = newdf.to_numpy()
# print(initial_prob_array)

initial_index_list = list(newdf.index.values)
# print(initial_index_list)
initial_df_sum
```

```python
transition_prob_dict = {}
# generate the transition probability matrix for each minute
# tabulated transition probabilities for one activity to the other, such as from work to traveling, 
# for every minute of the day. That provided 1,440 transition matrices, which let me model a day as 
# a time-varying Markov chain.
def transition_prob_matrix(minute):
    if minute == 1439:
        transition_prob = pd.crosstab(df_processed[0], df_processed[minute], normalize=1).reindex(columns=df_unique_activites, index=df_unique_activites, fill_value=0)
        transition_prob_dict[minute] = transition_prob.transpose()
        return transition_prob
    else: 
        transition_prob = pd.crosstab(df_processed[minute + 1], df_processed[minute], normalize=1).reindex(columns=df_unique_activites, index=df_unique_activites, fill_value=0)
        transition_prob_dict[minute] = transition_prob.transpose()
        return transition_prob

for i in range(0, 1440):
    transition_prob_matrix(i)
```

```python
def state_change(start_state, minute):
    # print(initial_index_list)
    # print(initial_prob_array)
    testing = np.arange(0, len(initial_index_list), 1).astype(int)
    # print(testing)
    testing_state = initial_index_list.index(start_state)
    # print(testing_state)
    row = transition_prob_dict[minute].iloc[testing_state]
    # print(row)
    end_index = np.random.choice(testing, p=row)
    # print(end_index)
    end_state = initial_index_list[end_index]
    # print(end_state)
    return end_state
    
```

```python
def activity_mc(n, history=True):
    # n is the number of minutes to simulate
    # history is a boolean that determines whether to return the full history of the simulation
    # or just the final state
    # initialize the simulation
    # the initial state is chosen from the initial probability distribution
    initial_state = np.random.choice(initial_index_list, p=initial_prob_array)
    # print(initial_state)
    # initialize the history of the simulation
    stateHistory = [initial_state]
    # initialize the current state
    start_state = initial_state
    # simulate the system for n minutes
    for i in range(n):
        for attempt in range(10):
            try:
                # do thing
                start_state = state_change(start_state, i)
            except:
                # perhaps reconnect, etc.
                print("Something went wrong")
            else:
                break
        if history:
            stateHistory.append(start_state)
    # return the history or the final state
    if history:
        return stateHistory
    else:
        return start_state
```

```python
def collection(n):
    collection = {}
    for i in range(n):
        personID = generate(size=10)
        collection[personID] = activity_mc(1440)
    return pd.DataFrame.from_dict(collection, orient ='index') 

final_data = collection(500)
```
